### 6.1 Fixing *engV* missing / erroneous values

 From the Data Information:                     
 1.There are 434 missing values in engV(variable) and 511 missing values in drive(variable)               
 2.Duplicate rows of 113 
 

```python
#Drop duplicate rows
sls_data.drop_duplicates(inplace=True)
sls_data.shape
```

Missing Values replacement

```python
#which method to use replace the missing values..is it mean, median or mode?How can we find out?
#By plotting distribution plots
sd = plt.subplots(figsize=(10,8))
sns.distplot(sls_data.engV)
```

From the distribution plot, we find out that it is highly skewed,hence we have to calculate mean for missing values in engV.   

```python
#Mean Value
sls_data_mean=sls_data['engV'].mean()
print(sls_data_mean)
```

```python
'''
Remove outlier Function
'''
def remove_outliers_engV():
    return pd.DataFrame(sls_data[(sls_data.engV > 0.5) & (sls_data.engV < 30)])


'''
Mean Value function engV based on brand and model 
'''
def mean_engV(sd):
    car_brand = sd.name[0]    
    car_model = sd.name[1]
    if sd.count() > 0:
        return round(sd.mean(),2)
    else:
        engV_sd = remove_outliers_engV()
        engV_brand = engV_sd[engV_sd.car == car_brand].engV
        if engV_brand.count() > 0:
            return round(engV_brand.mean(),2)
        else:
            return round(engV_sd.engV.mean(),2)
        
'''
Function for  all combinations of brand and model
'''
engV_mean = sls_data.groupby(['car','model']).engV.apply(mean_engV).reset_index()
engV_mean.rename(columns={'engV':'engV_avg'}, inplace=True)
engV_mean.head(5)
```

```python
'''
Replace EngV mean with missing values
'''
def missing_engV(sd):
    if pd.isnull(sd.engV):
        return engV_mean[(engV_mean.car == sd.car) & (engV_mean.model == sd.model)].engV_avg.values[0]
    else:
        return sd.engV
    
sls_data.engV = sls_data.apply(missing_engV, axis=1)
print('After Replacement of missing Values')
print(sls_data[sls_data.engV.isnull()].shape)
```

### 6.2 Fix missing *price*  / erroneous values

```python
'''Mean value cannot be used here in replacing the missing values as there are different car brands and models and hence median is
used'''
#Replace missing values and correct erroneous data in price

def cal_price_median(sd):
    car_brand = sd.name[0]    
    car_model = sd.name[1]
    if (sd.count() > 0) & (sd.median() > 0):
        return sd.median()
    else:
        price_make = sls_data[sls_data.car == car_brand].price
        if price_make.median() > 0:
            return price_make.median()
        else:
            return sls_data.price.median()

        
#Applying median values across car brands and models

price_median = sls_data.groupby(['car','model']).price.apply(cal_price_median).reset_index()
price_median.rename(columns={'price':'price_median'}, inplace=True)
price_median.head(5)
```

```python
#median price returns the missing values

def missing_price(sd):
    if sd.price == 0:
        return price_median[(price_median.car == sd.car) & (price_median.model == sd.model)].price_median.values[0]
    else:
        return sd.price
    
#Fix missing values

sls_data.price = sls_data.apply(missing_price, axis=1)
print(sls_data[sls_data.price == 0].shape)
```

### 6.3 Fixing *missing* mileage values / erroneous values

```python
'''Mean value cannot be used here in replacing the missing values as there are different car brands and models and hence median is
used'''

#Replace missing values and correct erroneous data in mileage

def cal_mileage_median(sd):
    car_brand = sd.name[0]    
    car_model = sd.name[1]
    if (sd.count() > 0) & (sd.median() > 0):
        return sd.median()
    else:
        mileage_make = sls_data[sls_data.car == car_brand].price
        if mileage_make.median() > 0:
            return mileage_make.median()
        else:
            return sls_data.mileage.median()

        
#Applying median values across car brands and models

mileage_median = sls_data.groupby(['car','model']).mileage.apply(cal_mileage_median).reset_index()
mileage_median.rename(columns={'mileage':'mileage_median'}, inplace=True)
mileage_median.head(5)
```

```python
#median mileage returns the missing values

def missing_mileage(sd):
    if sd.mileage == 0:
        return mileage_median[(mileage_median.car == sd.car) & (mileage_median.model == sd.model)].mileage_median.values[0]
    else:
        return sd.mileage
    
#Fix missing values

sls_data.mileage = sls_data.apply(missing_mileage, axis=1)
print(sls_data[sls_data.mileage == 0].shape)    
```

### 6.4 Fix missing *drive* values / erroneous values

```python
'''
Remove null entries..
'''
def cleaned_data_drive():
    return pd.DataFrame(sls_data[sls_data.drive.isnull() == False])

#correct erroneous data in drive


def New_drive(sd):
    car_brand = sd.name[0]    
    car_model = sd.name[1]
    if sd.count() > 0:
        return sd.mode()
    else:
        clndrive_df = cleaned_data_drive()
        drive_brand = clndrive_df[clndrive_df.car == car_brand].drive
        if drive_brand.count() > 0:
            return drive_brand.mode()
        else:
            return clndrive_df.drive.mode()
                

#Applying mode for all combinations in brands and models

drive_mode = sls_data.groupby(['car','model']).drive.apply(New_drive).reset_index()
drive_mode.rename(columns={'drive':'drive_mode'}, inplace=True)
drive_mode.drop('level_2', axis=1)
```

```python
#missing values in drive

def missing_drive(sd):
    if pd.isnull(sd.drive):
        return drive_mode[(drive_mode.car == sd.car) & (drive_mode.model == sd.model)].drive_mode.values[0]
    else:
        return sd.drive
    
#Fix missing values
sls_data.drive = sls_data.apply(missing_drive, axis=1)
print(sls_data[sls_data.drive.isnull()].shape) 
```

### 6.5 Fixing *engType* missing / erroneous values

```python
#correct erroneous data in engType
'''
Remove null entries..
'''
def cleaned_data_engType():
    return pd.DataFrame(sls_data[sls_data.engType.isnull() == False])

#correct erroneous data in drive


def New_engType(sd):
    car_brand = sd.name[0]    
    car_model = sd.name[1]
    if sd.count() > 0:
        return sd.mode()
    else:
        clnengType_df = cleaned_data_engType()
        engType_brand = clnengType_df[clnengType_df.car == car_brand].engType
        if engType_brand.count() > 0:
            return engType_brand.mode()
        else:
            return clnengType_df.engType.mode()
                

#Applying mode for all combinations in brands and models

engType_mode = sls_data.groupby(['car','model']).engType.apply(New_engType).reset_index()
engType_mode.rename(columns={'drive':'engType_mode'}, inplace=True)
engType_mode.drop('level_2', axis=1)
```

```python
'''
Missing values in engType
'''
def missing_engType(sd):
    if pd.isnull(sd.engType):
        return engType_mode[(engType_mode.car == sd.car) & (engType_mode.model == sd.model)].engType_mode.values[0]
    else:
        return sd.engType

sls_data.engType = sls_data.apply(missing_engType, axis=1)
print(sls_data[sls_data.engType.isnull()].shape)
```

### 8.5 Q - What are the sales & revenue in each segment of class ?

```python
#section of cars based on Price classfication:
#1.Economy = Price < 10000 
#2.Luxury  = Price b/w 10000 and 30000
#3.Premium = Price > 30000

def classif(sd):
    if sd.price <= 10000:
        return "Economy"
    elif (sd.price > 10000) & (sd.price <= 30000):
        return "Luxury"
    else:
        return "Premium"
```

```python
#Create the 'class' for classifying

sls_data["class"] = sls_data.apply(classif, axis=1)
```

```python
sd = sls_data["class"].value_counts().plot.bar(figsize=(6,4))
plt.title("Car sales based on Class")
```

```python
revenueByClass_df = sls_data.groupby('class').price.sum()
plt.title('Sales revenue based on Class')
print(revenueByClass_df.plot.pie(autopct='%1.1f%%'))
```

### 8.7 Q - Which engV has the maximum sales?

```python
#section of cars based on Price classfication:
#1.Less Power = cc < 2 
#2.Medium Power  = cc b/w 2 and 4
#3.High Power = cc > 4

def volume(sd):
    if sd.engV <= 2:
        return "Less Power"
    elif (sd.engV > 2) & (sd.engV <= 4):
        return "Medium Power"
    else:
        return "High Power"
```
